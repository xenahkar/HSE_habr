```python
fr = big_frame_only_text.copy()
```

```python
noise = stopwords.words("russian") + list(punctuation) # шум
```

####  текст статьи

```python
%%time
for i in tqdm(fr.index):
    fr['text'][i] = word_tokenize(fr['text'][i]) # разбиваем на токены
    t = fr['text'][i]
    t1 = [x for x in t if x not in noise] # убираем пунктуацию и стоп-слова
    fr['text'][i] = t1
```

```python
fr.head(3)
```

```python
from pymystem3 import Mystem

mystem_analyzer = Mystem()
t5 = mystem_analyzer.lemmatize(' '.join(str(i) for i in fr['text'][0]))
''.join(str(i) for i in t5)
```

```python
%%time
for i in tqdm(fr.index):
    t5 = mystem_analyzer.lemmatize(' '.join(str(k) for k in fr['text'][i])) # лемматизируем
    fr['text'][i] = ''.join(str(j) for j in t5)
```

#### название статьи

```python
col1 = 'title'

def text_clean(fr, col: str):
    for i in tqdm(fr.index):
        fr[col][i] = word_tokenize(str(fr[col][i])) # разбиваем на токены
        t = fr[col][i]
        t1 = [x for x in t if x not in noise] # убираем пунктуацию и стоп-слова
        t2 = mystem_analyzer.lemmatize(' '.join(str(j) for j in t1)) # лемматизируем
        t3 = ''.join(str(k) for k in t2)
        fr[col][i] = t3
    return fr

text_clean(fr, col1)
```

```python
%%time
fr['article'] = '' # создаем столбец с НАЗВАНИЕ СТАТЬИ + ТЕКСТ СТАТЬИ
for i in tqdm(fr.index):
    fr['article'][i] = fr['title'][i] + fr['text'][i]
```

#### Преобразуем хабы в ТОП-10

```python
list(fr['hubs_pro'])
```

```python
def hub_count(fr, hubs:str):
    _count = dict()

    for _list in fr[hubs]:
        for i in _list:
            if i not in _count:
                _count[i] = 1
            else:
                _count[i] += 1

    _count = {k: v for k, v in sorted(_count.items(), key=lambda item: item[1], reverse=True)}
    df = pd.DataFrame.from_dict(_count, orient='index', columns = ['count']).reset_index()
    return df

hub_count(fr, 'hubs_pro')
```
