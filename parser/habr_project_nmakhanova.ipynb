```python
def hubs_pars(soup):
    hubs = []
    hub = soup.find_all('a', {'class':"tm-publication-hub__link"})
    for i in range(len(hub)):
        subhub = hub[i].text.strip()
        if '*' in subhub:
            hubs.append(1)
            hubs.append(subhub.replace(' *', ''))
        else:
            hubs.append(subhub)
    return hubs
hubs_pars(soup)
```

```python
#доп. инфо
def text_pars(soup):
    text_full = ''
    text = soup.find_all('p')
    for i in range(len(text)):
        text_full = text_full + text[i].text
    return text_full
text_pars(soup)
```

```python
def links_pars(soup):
    links = []
    links_in_article = soup.find_all('a', {'rel': "noopener noreferrer nofollow"})
    for i in range(len(links_in_article)):
        links.append(links_in_article[i].get('href'))
    return links
links_pars(soup)
```

```python
def tags_pars(soup):
    tags = []
    all_tags = soup.find_all('a', {'class': "tm-tags-list__link"})
    for i in range(len(all_tags)):
        tags.append(all_tags[i].text)
    return tags

tags_pars(soup)
```

```python
def vote(soup):
    votes_class = "tm-votes-meter__value tm-votes-meter__value tm-votes-meter__value_negative tm-votes-meter__value_appearance-article tm-votes-meter__value_rating"
    if len(soup.find_all('span', {'class': votes_class})) != 0:
        vote_info = soup.find_all('span', {'class': votes_class})[0]
        vote_diff = int(vote_info.text)
        votes = soup.find_all('span', {'class': votes_class})[0].get('title')
        all_votes = [int(i) for i in re.findall(r'\d{1,}', votes)]
        all_votes.append(vote_diff)
        return all_votes
    else:
        return 0
vote(soup)
```
